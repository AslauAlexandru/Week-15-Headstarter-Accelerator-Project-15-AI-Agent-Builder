```python
# Multi-Modal Agentic Applications:
# https://cobusgreyling.medium.com/multi-modal-agentic-applications-4a15ab0c0bf6

### Image URL Setup
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"

#### Environment Setup
import os
from google.colab import userdata
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')

#### Model Initialization
from langchain_core.messages import HumanMessage
#from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI


model = ChatGoogleGenerativeAI(model="gemini-1.5-pro")
#model = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
#model = ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite")


### The httpx library fetches the image from the URL, and the base64
### library encodes the image data in base64 format.
### This conversion is necessary to send the image data as part of the
### input to the mode.

import base64
import httpx
image_data = base64.b64encode(httpx.get(image_url).content).decode("utf-8")


### Constructing the message. This creates a HumanMessage object that contains
### two types of input: a text prompt asking to describe the weather in the
### image, and the image itself in base64 format.
message = HumanMessage(
    content=[
        {"type": "text", "text": "Analyze product images and provide detailed descriptions"},
        {
            "type": "image_url",
            "image_url": {"url": f"data:image/jpeg;base64,{image_data}"},
        },
    ],
)


### The model.invoke method sends the message to the model,
###  which processes it and returns a response.
response = model.invoke([message])

### The result, which should be a description of the weather in the image,
### is then printed out.
print(response.content)
```

```python
def image_to_text(image_url):
    ### Image URL Setup
    #image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"

    #### Environment Setup
    import os
    from google.colab import userdata
    os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')

    #### Model Initialization
    from langchain_core.messages import HumanMessage
    #from langchain_openai import ChatOpenAI
    from langchain_google_genai import ChatGoogleGenerativeAI


    model = ChatGoogleGenerativeAI(model="gemini-1.5-pro")
    #model = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
    #model = ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite")


    ### The httpx library fetches the image from the URL, and the base64
    ### library encodes the image data in base64 format.
    ### This conversion is necessary to send the image data as part of the
    ### input to the mode.

    import base64
    import httpx
    image_data = base64.b64encode(httpx.get(image_url).content).decode("utf-8")


    ### Constructing the message. This creates a HumanMessage object that contains
    ### two types of input: a text prompt asking to describe the weather in the
    ### image, and the image itself in base64 format.
    message = HumanMessage(
        content=[
            {"type": "text", "text": "Analyze product images and provide detailed descriptions"},
            {
                "type": "image_url",
                "image_url": {"url": f"data:image/jpeg;base64,{image_data}"},
            },
        ],
    )


    ### The model.invoke method sends the message to the model,
    ###  which processes it and returns a response.
    try:
      response = model.invoke([message])

      ### The result, which should be a description of the weather in the image,
      ### is then printed out.
      return response.content

    except Exception as e:  # Catch any exception
        print(f"Error processing image: {e}")  # Print the error message
        return "Error processing image. Please check the URL or try again later. In any case URL or link must be end in .jpg or in .png or in other format for image." # Return an error message to the user



# Import the gradio library
import gradio as gr

interface = gr.Interface(
    fn=image_to_text,
    inputs=gr.Textbox(label="Your url or link to image", placeholder="Type your question here and press Enter..."),
    #outputs=gr.Textbox(label="Response"), # Define the outputs here
    outputs=gr.Markdown(label="Response"),
    title="Gemini Chat AI Agent Application (AI Agentic Applications with Langchain)",
    description="Provide the url or the link to image for the Gemini LLM and receive responses after pressing Submit.",
    examples=[["https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"]] # Changed to a list containing example input
)

interface.launch(debug=True)
```
